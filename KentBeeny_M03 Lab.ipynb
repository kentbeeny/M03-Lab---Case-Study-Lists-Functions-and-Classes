```python
#M03 Lab - Case Study: Lists, Functions, and Classes
#Kent Beeny
#SDEV 220
#11/14/2023

#the vehicle super class
class Vehicle:
    def __init__(self, Type):
        self.vehType = Type
        
#the Automobile child class
class Automobile(Vehicle):
    def __init__(self, Type):
        super().__init__(Type)
        self.year = 1111
        self.make = ""
        self.model = ""
        self.doors = ""
        self.roof = ""
        
def main():
    Type = input("What type of automobile is this?: \n")
    
    #get the Year of the automobile
    while True:
        try:
            Year = int(input("What year was your Automobile produced?: "))
            if Year < 1:
                print("That's way too old of a car, please enter again.")
            elif Year > 2024:
                print("You can't have a car from the future, please enter again.")
            else:
                break
        except ValueError:
            print("That is not a number, please try again.")
            
    #get the make of the automobile
    Make = input("What make is your Automobile?: ")
    #get the model of the automobile
    Model = input("What model is your automobile?: ")
    #get the number of doors on the automobile
    while True:
        try:
            Doors = int(input("How many doors does your automobile have \(2 or 4\)?: "))
            if Doors == 2 or Doors == 4: 
                break
            else:
                print("You've entered an invalid number, please try again.")
        except ValueError:
            print("That isn't a number, try again.")\
            
    #get the roof of the automobile
    while True:
            Roof = "it has one"
            getRoof = input("How style roof does your automobile have \('sun' for sunroof or 'solid' for a solid roof\)?: ")
            if getRoof == 'sun':
                Roof = "Sun Roof"
                break
            elif getRoof == 'solid':
                Roof == 'Solid Roof' 
                break
            else:
                print("You've entered an invalid roof type, please try again.")
    myVehicle = Automobile(Type)
    myVehicle.Type = Type
    myVehicle.Year = Year
    myVehicle.Make = Make
    myVehicle.Model = Model
    myVehicle.Doors = Doors
    myVehicle.Roof = Roof
    
    print("Vehicle Type: " + str(myVehicle.Type))
    print("Vehicle Year: " + str(myVehicle.Year))
    print("Vehicle Make: " + str(myVehicle.Make))
    print("Vehicle Model: " + str(myVehicle.Model))
    print("Vehicle Doors: " + str(myVehicle.Doors))
    print("Vehicle Roof: " + str(myVehicle.Roof))
    
    
main()
```
